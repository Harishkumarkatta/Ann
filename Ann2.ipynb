```python
from tensorflow import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras.models import Sequential
```

```python
### List columns names based on the description
col_names = ['class', 'lymphatics', 'block of affere', 'bl. of lymph. c', 'bl. of lymph. s', 'by pass', 
 'extravasates', 'regeneration of', 'early uptake in', 'lym.nodes dimin', 'lym.nodes enlar', 
'changes in lym.', 'defect in node', 'changes in node', 'changes in stru', 'special forms', 
'dislocation of', 'exclusion of no', 'no. of nodes in']
data=pd.read_csv("lymphography.data",names=col_names)
data.head()
```

```python
data["class"].value_counts() # this is dependent variable
```

```python
data.describe()
```

```python
#independent variable
y=data["class"]
# independent variable
X=data.drop("class",axis=1)
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
# our class has imbalanced data we use
!pip install imblearn
```

```python
df_train=pd.concat([X_train,Y_train],axis=1)
```

```python
df_train["class"].value_counts()
```

```python
from sklearn.utils import resample
df_train_up = df_train[df_train["class"]==2]
for n in [1, 3, 4]:
    upsampled = resample(df_train[df_train["class"]==n],
                        replace=True, # sample with replacement
                        n_samples=59, # match number in majority class
                        random_state=1) # reproducible results
    df_train_up = pd.concat([df_train_up, upsampled])
```

```python
df_train_up["class"].value_counts()
```

```python
upsampled.head()
```

```python
x_train_up = df_train_up.drop(["class"], axis=1)
y_train_up = df_train_up["class"]
```

```python
x_train_up.shape
```

```python
y_train_up .shape
```

```python
from keras.layers import Dense
#
classifier=Sequential() # empty neural netwwork
#first hidden layer
classifier.add(Dense(units=1000,kernel_initializer="he_uniform",activation="relu",input_dim=18))# 6 hidden neuron,init=wt initialization
#second hidden layer
classifier.add(Dense(units=512,kernel_initializer="he_uniform",activation="relu")) # 18 input layers
# output layer given 1 beacause it is only binary classification
classifier.add(Dense(units=5,kernel_initializer="he_uniform",activation="sigmoid"))
```

```python
classifier.summary()
```

```python
from tensorflow.keras.losses import SparseCategoricalCrossentropy
classifier.compile(optimizer="adam",loss="SparseCategoricalCrossentropy",metrics=["accuracy"])
```

```python
model_history=classifier.fit(x_train_up,y_train_up,validation_split=0.33,batch_size=10,epochs=10)
```
